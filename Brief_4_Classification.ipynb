## 1.3 Répartition des données

```python
# Répartition de la base d'entraînement et de la base de test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 1/7 , random_state=42)
```

## 2  Apprentissage d’un classifieur binaire

### 2.1 Apprentissage des données

### 2.1.1

```python
# On crée une liste de booléen pour savoir si la cible est égale à la première instance(5) ou pas :
 
ytrain5 = Y_train == 5

ytrain5
```

### 2.1.2

### 2.2 Évaluation du modèle d’apprentissage sur les données d’apprentissage

```python
# Mise en instance de ma classe SGDClassifier
import numpy as np

from sklearn.linear_model import SGDClassifier
sgdc = SGDClassifier(max_iter=5, tol=-np.infty)
```

```python
# Entraînement du modèle
sgdc.fit(X_train,ytrain5)
```

### 2.2.1 Taux de classification

### 2.2.1.1

```python
# Validation croisée 1 pour le taux de classification de chaque fold
from sklearn.model_selection import cross_val_score
cvs = cross_val_score(sgdc, X_train, ytrain5,scoring ="accuracy", cv = 3)
```

```python
# La moyenne des taux de classification
cvs.mean()
```

### 2.2.2 Matrix de confusion

```python
#Fonction cross_val_predict
from sklearn.model_selection import cross_val_predict
Y_train_predict = cross_val_predict(sgdc, X_train,ytrain5, cv=3)
```

```python
# Matrice de confusion
from sklearn.metrics import confusion_matrix
confusion_matrix(ytrain5,Y_train_predict)
```

### 2.2.3 Précision et rappel:


```python
# La précision
from sklearn.metrics import precision_score
precision_score(ytrain5,Y_train_predict)
```

```python
# Rappel
from sklearn.metrics import  recall_score
recall_score(ytrain5,Y_train_predict)
```

```python
# F1_score
from sklearn.metrics import f1_score
f1_score(ytrain5,Y_train_predict)
```

#### 2.2.3.2

```python
# Le classifieur SGD calcule un score en se basant sur sa fonction de décision. Si le score est supérieur à un seuil, il affecte la classe positive à l’instance
# sinon il affecte la classe négative.
from sklearn.model_selection import cross_val_predict
scores1 = cross_val_predict(sgdc, X_train,ytrain5, method= 'decision_function',cv=3)
scores1
```

#### 2.2.3.3

```python
# Précisions et les rappels ( et seuils)
from sklearn.metrics import precision_recall_curve
precision, recall, thresholds = precision_recall_curve(ytrain5, scores1)
```

#### 2.2.3.4

```python
# Courbe des précisions/rappels
plt.plot(recall[:-1], precision[:-1])
plt.ylabel("Précisions")
plt.xlabel("Rappels")
plt.show()
```

La courbe commence à sérieusement diminuer entre 60 % et 80 % . 

### 2.2.4 Courbe ROC

#### 2.2.4.1

```python
# le taux de faux positifs (tfp), le taux de vrais positifs (tvp)
from sklearn.metrics import roc_curve
tfp, tvp, thresholds = roc_curve(ytrain5, scores1)
```

#### 2.2.4.2

```python
plt.ylabel("tvp")
plt.xlabel("tfp")
plt.plot(tfp,tvp,color='r')
plt.show()
```

#### 2.2.4.3

```python
# Evaluation de la courbe ROC
from sklearn.metrics import roc_auc_score
roc_auc_score(ytrain5,scores1)
```
